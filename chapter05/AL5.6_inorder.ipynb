>이진트리의 중위 순회

>#### **2. 알고리즘 설명(MD)  (알고리즘 설명)**
>중위 순회는 LVR로 이진트리의 왼쪽부터 가운데를 들려 오른쪽으로 순차적으로 본다

#### **3. 손으로 푼 예제 (![이름][img]/이미지.jpg])  (책 예제 코드 이해 후 사진)**

>#### **4. 코드 개요(입력 변수, 출력, 알고리즘 개요) (MD)**
>##### (코드 개요 설명 1. 입력 변수 2. 출력 3. 함수설명)
>1. 입력 변수 : inorder(n)
>2. 출력      &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;: 호출문으로 호출을 할 시 가장 왼쪽 서브 트리부터 처리를 하여 순차적으로 오른쪽 서브 트리까지 처리한다
>3. 함수 설명 : 왼쪽을 먼저 처리를 진행 후에 오른쪽으로가는데 순차적으로 기억을 해둔 후 출력을 해주는 함수이다


---------

#### **5. 코드(PY, C)**

```python
class TNode:
    def __init__(self, data, left, right):
        self.data = data
        self.left = left
        self.right = right

def inorder(n):
    if n is not None :
        inorder(n.left)
        print(n.data, end=' ')
        inorder(n.right)
        

d = TNode('D', None, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', None, None)
c = TNode('C', f, None)
root = TNode('A', b, c)

print(' In-Order : ', end=' ')
inorder(root)
print()
        
```

#### **6. 테스트 코드(PY, C), 입력을 바꾸어서**

```python
class TNode:
    def __init__(self, data, left, right):
        self.data = data
        self.left = left
        self.right = right

def inorder(n):
    if n is not None :
        inorder(n.left)
        print(n.data, end=' ')
        inorder(n.right)
        

z = TNode('Z', None, None)
i = TNode('I', None, z)
h = TNode('H', None, None)
p = TNode('P', None, None)
g = TNode('G', None, p)
d = TNode('D', g, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', h, i)
c = TNode('C', f, None)
root = TNode('A', b, c)

print(' In-Order : ', end=' ')
inorder(root)
print()
        
```
