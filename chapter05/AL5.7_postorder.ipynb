>#### **4. 코드 개요(입력 변수, 출력, 알고리즘 개요) (MD)**
>##### (코드 개요 설명 1. 입력 변수 2. 출력 3. 함수설명)
>1. 입력 변수 : postorder(n)
>2. 출력      &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;: 호출문으로 호출을 할 시 가장 왼쪽 서브 트리부터 처리를 하여 순차적으로 오른쪽 서브 트리까지 처리한다.
>3. 함수 설명 : 중위 수식과 다른점은 중위 수식은 가장 왼쪽부터 Y축으로 순서대로 읽는다면 후위 수식은 왼쪽의 자식노드부터 읽어 올라가며 오른쪽도 똑같이 반복한다

---------

#### **5. 코드(PY, C)**

```python
class TNode:
    def __init__(self, data, left, right):
        self.data = data
        self.left = left
        self.right = right

def postorder(n):
    if n is not None :
        postorder(n.left)
        postorder(n.right)
        print(n.data, end=' ')
        

d = TNode('D', None, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', None, None)
c = TNode('C', f, None)
root = TNode('A', b, c)

print(' In-Order : ', end=' ')
postorder(root)
print()
        
```

#### **6. 테스트 코드(PY, C), 입력을 바꾸어서**

```python
class TNode:
    def __init__(self, data, left, right):
        self.data = data
        self.left = left
        self.right = right

def postorder(n):
    if n is not None :
        postorder(n.left)
        postorder(n.right)
        print(n.data, end=' ')

z = TNode('Z', None, None)
i = TNode('I', None, z)
h = TNode('H', None, None)
p = TNode('P', None, None)
g = TNode('G', None, p)
d = TNode('D', g, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', h, i)
c = TNode('C', f, None)
root = TNode('A', b, c)

print(' In-Order : ', end=' ')
postorder(root)
print()
```
