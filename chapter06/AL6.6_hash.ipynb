# 문제 정의

선형 조사형의 삽입 알고리즘

# 알고리즘 설명
1. 선형 조사법은 해시 테이블에서 충돌을 해결하는 방법이다.
2. 충돌이 발생했을 때, 다음 해시 슬롯을 탐색하여 빈 슬롯을 찾고 빈 공간이 있으면 저장한다.
3. 비어있는 공간이 나올 때까지 계속한다.
4. 테이블의 끝에 도달하면 다시 테이블의 처음으로 돌아간다.


# 손으로 푼 예제

# 코드 개요

1. 입력 변수 : key는 삽입하는 키다.
2. 출력 :
3. 함수 설명 : 1. hashFn 함수를 사용하여 키의 초기 해시 값을 계산한다.
              2. 해시 테이블 크기인 M을 count 변수에 저장한다.
              3. 반복문을 사용하여 빈 슬롯을 찾을 때까지 선형 조사법으로 슬롯을 탐색한다.

# 알고리즘 코드

```python
M = 13
table = [None]*M
def hashFn(key) :
    return key % M

def lp_insert(key) :
    id = hashFn(key)
    count = M
    while count>0 and (table[id] != None) :
            id = (id + 1 + M) % M
            count -= 1
            if count > 0 :
                table[id] = key
            return
```

# 테스트 코드

```python
M = 13
table = [None] * M

def hashFn(key):
    return key % M

def lp_insert(key):
    id = hashFn(key)
    count = M
    while count > 0:
        if table[id] is None:
            table[id] = key
            return
        id = (id + 1) % M
        count -= 1
    print("Hash table is full, could not insert key:", key)


def test_lp_insert():
    keys_to_insert = [45, 27, 88, 9, 71, 60, 46, 38, 24]
    for key in keys_to_insert:
        lp_insert(key)

    print("해시 테이블 상태:")
    for i, value in enumerate(table):
        print(f"Index {i}: {value}")

if __name__ == "__main__":
    test_lp_insert()
```

2. 공간 복잡도 함수는 : O(M)
