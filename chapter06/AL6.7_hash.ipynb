3. 해당 키의 레코드를 찾거나, 레코드가 없는 버킷을 만나거나 모든 버킷을 다 검사할 때까지 진행된다.

# 알고리즘 코드

```python
def lp_search(key) :
    id = hashFn(key)
    count = M
    while count>0 :
        if table[id] == None :
            return None
        if table[id] == key :
            return table[id]
        id = (id + 1 + M) % M
        count -= 1
    return None
```

# 테스트 코드

```python
M = 13
table = [None] * M

def hashFn(key):
    return key % M

def lp_insert(key):
    id = hashFn(key)
    count = M
    while count > 0:
        if table[id] is None or table[id] == "DELETED":
            table[id] = key
            return
        id = (id + 1) % M
        count -= 1
    print("Hash table is full, could not insert key:", key)

def lp_search(key):
    id = hashFn(key)
    count = M
    while count > 0:
        if table[id] is None:
            return None
        if table[id] == key:
            return table[id]
        id = (id + 1) % M
        count -= 1
    return None


def test_lp_insert_and_search():
    keys_to_insert = [38, 27, 45, 71, 60, 9, 88, 46, 24]
    
    for key in keys_to_insert:
        lp_insert(key)
    
    print("해시 테이블 상태:")
    for i, value in enumerate(table):
        print(f"Index {i}: {value}")

    keys_to_search = [38, 27, 45, 71, 60, 9, 88, 46 ,24]
    for key in keys_to_search:
        result = lp_search(key)
        if result is not None:
            print(f"Key {key} found: {result}")
        else:
            print(f"Key {key} not found.")

if __name__ == "__main__":
    test_lp_insert_and_search()
```
