# 알고리즘 설명 

주어진 동전의 종류를 사용하여 특정 금액을 만드는 데 필요한 최소 동전 수를 찾는 문제를 해결하기 위한 탐욕적 기법(greedy algorithm)이다. 단계에서 가장 큰 단위의 동전을 가능한 많이 사용하는 방식으로 진행된다.

# 코드 개요

1. 입력 변수 : 'coins' : 사용 가능한 동전의 종류를 나타내는 리스트. 예제에서는 '[400, 200, 100, 500, 50, 10]으로 주어진다.

2. 츌력 변수 : 'count' 각 동전 종류별로 사용된 동전의 개수를 담은 리스트. 각 동전의 순서와 일치하게 값을 반환한다.

3. 함수 설명 : for i in range(len(coins)): 동전 리스트의 각 요소에 대해 반복한다.
count.append(cnt): 계산된 동전의 개수를 count 리스트에 추가한다.
return count: 각 동전 종류별로 사용된 개수를 담고 있는 count 리스트를 반환한다.

# 손으로 푼 예제

# 알고리즘 코드

```python
def min_coins_greedy( coins, V ):
    count = []
    for i in range(len(coins)) :
        cnt = V // coins[i]
        count.append(cnt)
        V -= cnt * coins[i]
    return count
```

# 테스트 코드

```python
coins = [400 , 200, 100, 500, 50, 10]
V = 580
print("잔돈 액수 = ", V)
print("동전 종류 = ", coins)
print("동전 개수= ", min_coins_greedy(coins, V))
```

# 복잡도 분석

3행에서 동전의 종류만큼 루프를 반복하므로 동전의 액면가가 m가지라면 시간 복잡도는 O(m)이다.
