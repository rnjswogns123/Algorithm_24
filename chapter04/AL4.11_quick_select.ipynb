#### 1. 문제 정의(MD) (예제 그냥 문제 제목)

        축소 정복 기법을 이용한 k번째 작은 수 찾기

#### 2. 알고리즘 설명(MD)  (알고리즘 설명)


                k번째 작은 수를 찾으려면
                피벗으로 지정한 수가 인덱스+1한 수와 같은지를 판단하고 
                피벗 인덱스의 수가 작다면 답은 왼쪽에 있고 인덱스 + k번째 수를 찾는다.
                피벗 인덱스의 수가 크다면 답은 오른쪽에 있고 k-(인덱스+left)번째 수를 찾는다.

>3. 함수 설명 : 이 함수는 Quick Select 알고리즘을 사용하여 배열 A에서 k번째로 작은 값을 찾습니다. 함수는 배열 A와 배열의 왼쪽 경계 left, 오른쪽 경계 right, 그리고 찾고자 하는 k 값을 인수로 받습니다.  
&emsp;&emsp;&emsp;&emsp;&emsp; 함수는 재귀적으로 호출되며, 배열을 피벗을 기준으로 분할한 후 k번째 작은 값을 찾습니다. 그리고 해당 값을 반환합니다.

--------------------------

#### 5. 코드(PY, C)

```python
#4.11 k번째 작은수 찾기
def quick_select(A, left, right, k):
    pos = partition(A, left, right)
    
    if (pos+1 == left+k):
        return A[pos]
    elif(pos+1 > left+k):
        return quick_select(A, left, pos-1,k)
    else:
        return quick_select(A, pos+1,right,k-(pos+1-left))

#4.12 파티션
def partition(A, left, right) :  
    low = left + 1
    high = right
    pivot = A[left]
    while (low <= high) :
        while low <= right and A[low] <= pivot : low += 1
        while high >= left and A[high]> pivot : high-= 1
        
        if low < high :
            A[low], A[high] = A[high], A[low]
            
    A[left], A[high] = A[high], A[left]
    return high

array = [12, 3, 5, 7, 4, 19, 26, 23, 15]
print("입력 리스트= ",array)
n = len(array)
print("[축소기법] 3번째 작은 수:", quick_select(array, 0, n-1, 3))
print("[축소기법] 6번째 작은 수:", quick_select(array, 0, n-1, 6))
```
